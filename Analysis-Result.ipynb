# 4. Marketing ROI

CPC = Cost Per Click (How much you pay for each click)<br>
$$
CPC = \frac{Spend}{Clicks}
$$

CPA = Cost Per Acquistion (How much costs to generate one order/customer) <br>
$$
CPA = \frac{Spend}{Conversions}
$$

CTR = Click Through Rate (How many people clicked after seeing the ad)
$$
CTR = \frac{Clicks}{Impression}
$$

CVR = Conversion Rate (How many people clicks become customer)
$$
CVR = \frac{Conversions}{Clicks}
$$

RPC = Revenue Per Clicks (How many revenue each click generates)
$$
RPC = \frac{Revenue}{Clicks}
$$

ROAS = Return On Ad Spend (Every 1 dollar generates 4 dollars revenue)
$$
ROAS = \frac{Revenue}{Spend}
$$


```python
df_marketing_spend
```

```python
df_marketing_spend_platform = (df_marketing_spend
                               .groupby('platform', dropna=False, as_index=False)
                              .agg(
                                spend = ('spend','sum'),
                                impressions = ('impressions','sum'),
                                clicks = ('clicks','sum'),
                                conversions = ('conversions','sum'),
                                revenue_attributed = ('revenue_attributed','sum'),
                              )
                              )

df_marketing_spend_platform['ROAS'] =np.where(df_marketing_spend_platform['spend'] > 0 ,
                                             (df_marketing_spend_platform['revenue_attributed'] / df_marketing_spend_platform['spend']).round(2),
                                             np.nan)
df_marketing_spend_platform['CPC'] =np.where(df_marketing_spend_platform['clicks'] > 0 ,
                                             (df_marketing_spend_platform['spend'] / df_marketing_spend_platform['clicks']).round(2),
                                             np.nan)
df_marketing_spend_platform['CPA'] =np.where(df_marketing_spend_platform['conversions'] > 0 ,
                                             (df_marketing_spend_platform['spend'] / df_marketing_spend_platform['conversions']).round(2),
                                             np.nan)
df_marketing_spend_platform['CTR'] =np.where(df_marketing_spend_platform['impressions'] > 0 ,
                                             (df_marketing_spend_platform['spend'] / df_marketing_spend_platform['impressions']).round(3),
                                             np.nan)
df_marketing_spend_platform['CVR'] =np.where(df_marketing_spend_platform['clicks'] > 0 ,
                                             (df_marketing_spend_platform['conversions'] / df_marketing_spend_platform['clicks']).round(4),
                                             np.nan)
df_marketing_spend_platform['RPC'] =np.where(df_marketing_spend_platform['spend'] > 0 ,
                                             (df_marketing_spend_platform['revenue_attributed'] / df_marketing_spend_platform['spend']).round(2),
                                             np.nan)

df_marketing_spend_platform.sort_values('spend', ascending = False)
```

```python
df_marketing_spend_platform['ctr'] = np.where(df_marketing_spend_platform['impressions'] > 0, 
                                             df_marketing_spend_platform['clicks']/df_marketing_spend_platform['impressions'],
                                              np.nan)
df_marketing_spend_platform['cvr'] = np.where(df_marketing_spend_platform['clicks'] > 0, 
                                             df_marketing_spend_platform['conversions']/df_marketing_spend_platform['clicks'],
                                              np.nan)
df_marketing_spend_platform['rpc'] = np.where(df_marketing_spend_platform['clicks'] > 0, 
                                             df_marketing_spend_platform['revenue_attributed']/df_marketing_spend_platform['clicks'],
                                              np.nan)
```

```python
df_marketing_spend_platform.sort_values('spend', ascending = False)
```

```python
import numpy as np
import pandas as pd

df_marketing_spend['month'] = pd.to_datetime(df_marketing_spend['month'])

df_marketing_monthly = (
    df_marketing_spend
    .groupby(['month', 'platform'], as_index=False)
    .agg(
        spend=('spend', 'sum'),
        impressions=('impressions', 'sum'),
        clicks=('clicks', 'sum'),
        conversions=('conversions', 'sum'),
        revenue_attributed=('revenue_attributed', 'sum')
    )
)

df_marketing_monthly['cpc'] = np.where(
    df_marketing_monthly['clicks'] > 0,
    df_marketing_monthly['spend'] / df_marketing_monthly['clicks'],
    np.nan
)

df_marketing_monthly['cpa'] = np.where(
    df_marketing_monthly['conversions'] > 0,
    df_marketing_monthly['spend'] / df_marketing_monthly['conversions'],
    np.nan
)

df_marketing_monthly['roas'] = np.where(
    df_marketing_monthly['spend'] > 0,
    df_marketing_monthly['revenue_attributed'] / df_marketing_monthly['spend'],
    np.nan
)

df_marketing_monthly['ctr'] = np.where(
    df_marketing_monthly['impressions'] > 0,
    df_marketing_monthly['clicks'] / df_marketing_monthly['impressions'],
    np.nan
)

df_marketing_monthly['cvr'] = np.where(
    df_marketing_monthly['clicks'] > 0,
    df_marketing_monthly['conversions'] / df_marketing_monthly['clicks'],
    np.nan
)

df_marketing_monthly['rpc'] = np.where(
    df_marketing_monthly['clicks'] > 0,
    df_marketing_monthly['revenue_attributed'] / df_marketing_monthly['clicks'],
    np.nan
)

df_marketing_monthly
```

```python
df_roas_plot = df_marketing_monthly.pivot(
    index='month',
    columns='platform',
    values='roas'
)

df_roas_plot
```

5. If the CEO asked you to cut 20% of the marketing budget, which platforms and months would you recommend reducing spend on? Support your recommendation with data.

```python
df_marketing_spend_platform.sort_values('ROAS', ascending = False)
```

```python
#All Spend
print(f"All Spend = {df_marketing_spend_platform['spend'].sum()}")

spend_cut = df_marketing_spend_platform['spend'].sum() * 0.2
print(f"20% Spend Cut = {spend_cut}")

print(f"20% of 20% spend cut = {(spend_cut * 0.2).round(2)}")
print(f"35% of 20% spend cut = {(spend_cut * 0.35).round(2)}")
print(f"45% of 20% spend cut = {(spend_cut * 0.45).round(2)}")
```

```python
# Platform Spen after 20% cut
new_cut_platform = df_marketing_spend_platform.copy()

new_cut_platform.loc[new_cut_platform['platform'] == 'Email Marketing', 'spend'] = new_cut_platform.loc[new_cut_platform['platform'] == 'Email Marketing', 'spend'] - (spend_cut * 0.2) 
new_cut_platform.loc[new_cut_platform['platform'] == 'Facebook Ads', 'spend'] = new_cut_platform.loc[new_cut_platform['platform'] == 'Facebook Ads', 'spend'] - (spend_cut * 0.35) 
new_cut_platform.loc[new_cut_platform['platform'] == 'Google Ads', 'spend'] = new_cut_platform.loc[new_cut_platform['platform'] == 'Google Ads', 'spend'] - (spend_cut * 0.45) 

new_cut_platform.sort_values('ROAS', ascending = False)
```

answer: To meet the CEO’s request to reduce the marketing budget by 20%, the total marketing spend was first calculated and the required reduction amount was determined as 20% of the total spend. This reduction was then distributed across selected platforms based on their performance and budget size. Email Marketing receives 20% of the total reduction because, despite its lower ROAS, it demonstrates strong engagement metrics such as high RPC, CTR, and conversion rates, suggesting potential for optimization rather than elimination. The majority of the reduction is allocated to Facebook Ads (35%) and Google Ads (45%), which represent larger portions of the marketing budget and deliver moderate performance relative to top-performing platforms such as TikTok Ads and Influencer marketing. This strategy allows the company to achieve the required budget reduction while preserving the highest-performing channels and maintaining diversified marketing coverage.

```python
new_spend = []
total_roas = new_cut_platform['ROAS'].sum()
total_spend = new_cut_platform['spend'].sum()
for index, roas in enumerate(new_cut_platform['ROAS']):
    x = roas / total_roas * total_spend
    new_spend.append(x.round(2))

new_spend
```

```python
new_cut_platform['new_spend'] = new_spend
```
